0. Visualize the network

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pylab
import networkx as nx

import pickle
def pickle_load(path):
    with open(path, mode='rb') as f:
        data = pickle.load(f)
        return data

df = pd.read_csv("Tokyo_Edgelist.csv")
G = nx.read_edgelist("edges_2.txt", nodetype=int)
pos = pickle_load("positionDict.pickle")
pos_ = pos.copy()
#nx.draw(G, pos, node_size = 1, node_color = "b")
pos_
```

```python
## Select delivery place
import random
DELIV_NUM = 30
DELIV_PLACES = {}
posList = list(pos) # indexing key
for i in range(DELIV_NUM):
    node = random.choice(posList)
    cor = pos[node]
    DELIV_PLACES[node] = cor
DELIV_PLACES
```

```python
#nx.draw(G, pos_, node_size = 1, node_color = "b")
nx.draw(G, DELIV_PLACES, node_size = 1, node_color = "r")
plt.show()
```
