```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pylab
import networkx as nx

import pickle
def pickle_load(path):
    with open(path, mode='rb') as f:
        data = pickle.load(f)
        return data

df = pd.read_csv("networkData.csv")
pos = pickle_load("positionDict.pickle")
```

```python
## 続くダイクストラのために点をRenameしておく
df1, df2, df3 = df["START_NODE"], df["END_NODE"], df["LENGTH"]
#range(len(list(pos)))
df2 = df2.replace(list(pos),range(len(list(pos))))
df1 = df1.replace(list(pos),range(len(list(pos))))
edges = np.concatenate([np.array([df1]), np.array([df2]), np.array([df3])], axis = 0).T
np.savetxt("edge_weight.txt", edges, fmt = '%d')
G2 = nx.read_weighted_edgelist("edge_weight.txt", nodetype = int)
coords = df.values[:,1:3]
posRename = {}
# ndarrayの情報を「頂点:(x,y)」という値(辞書)に変換する
for i in range(len(coords) ):
    if i != 0 and (coords[i][0], coords[i][1]) == (coords[i-1][0], coords[i-1][1]):
        pass
    else:
        posRename[int(df1[i])] = (coords[i][0], coords[i][1])
#posRename
pos2 = posRename.copy()
```

```python
## Select delivery place
import random
DELIV_NUM = 40
DELIV_PLACES = {}
posList = list(posRename) # 辞書はリストじゃなくてセットオブジェクトだからindexing とかできない！！ので、こうやってkeyを取り出す必要あり
for i in range(DELIV_NUM):
    node = random.choice(posList)
    cor = posRename[node]
    DELIV_PLACES[node] = cor
#DELIV_PLACES
G1 = nx.empty_graph(0)
G1.add_nodes_from(list(DELIV_PLACES))
```

```python
nx.draw(G2, posRename, node_size = 1, node_color = "b")
nx.draw(G1, DELIV_PLACES, node_size = 10, node_color = "r")
plt.show()
```

●続いて配送計画問題

```python
droneNum = 4
#nx.draw(G1, DELIV_PLACES, node_size = 10, node_color = "r")
orderedNames = list(DELIV_PLACES.keys())
N_CLUST = int(len(orderedNames) / droneNum)
coors =np.array([list(DELIV_PLACES[i]) for i in orderedNames])
plt.scatter(coors[:,0], coors[:,1], marker = ".", c="red", s = 20 )
plt.show()
```

# 単純に近い点をクラスタリングするのではうまくいかない？？
→まず、クラスター内の点の個数を指定した方がいいかもしれない
However, 個数を指定する方法については結構色々とあるっぽい。。https://stackoverflow.com/questions/5452576/k-means-algorithm-variation-with-equal-cluster-size

```python
from samesizeKmeans import EqualGroupsKMeans
clf = EqualGroupsKMeans(n_clusters = N_CLUST)
clf.fit(coors)
labels = clf.labels_
np.savetxt("clusters.txt",labels, fmt = '%d')
```

```python
cmap = plt.get_cmap("tab10") 
cl = ["r", "b", "g","orange", "brown", "violet", "black"]
centroids = np.zeros((0,2))
for cluster in range(N_CLUST):
    data_ = coors[np.where(labels == cluster)]
    centroid_ = np.mean(data_, axis = 0)
    centroids = np.append(centroids, np.array([centroid_]), axis = 0)
    plt.scatter(data_[:,0], data_[:,1], marker = ".", c =cl[int(cluster%7)], s = 40)
    plt.scatter(centroid_[0], centroid_[1], marker = "*", c =cl[int(cluster%7)], s = 60, alpha = 0.3)
plt.show()
print(centroids)
```

# 次はそれぞれのクラスターの重心に近い点を探してきて車がそこを通る
→とりあえず重心を出して(このときに次数で候補絞るとかは要検討)

重心のリストが出たらあとはそこを回るだけ


```python
#coords = np.loadtxt("coords.csv", delimiter = ",")
cords = np.array([pos2[key] for key in pos2.keys()])

def dist(a,b):
    return np.sqrt(np.dot(b-a, b-a))

centerNodes = np.zeros(0)
for i in range(int(len(centroids[:,0])) ):
    centroid_ = centroids[i]
    distBox = [dist(cords[j,:], centroid_) for j in range(int(len(cords[:,0])))]
    dist_ = np.array(distBox)
    #print(np.sum(dist_))
    centerNodes = np.append(centerNodes, np.argmin(dist_) )
print(centerNodes)
#print(pos2.keys())
```

# 続いてここで得られた点を回るルートを作る
→このルート選びは人間の目には簡単だけど機械にはむずい？！

●一旦サイクリックヒューリスティクスなるものを考えてみる→すべての点との距離が最小になる円(最小二乗円)を作ってその偏角の順に回っていく。
https://github.com/bdhammel/least-squares-ellipse-fitting

```python
import ellipses as el
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

data = el.make_test_ellipse()
#print(data)

lsqe = el.LSqEllipse()
lsqe.fit([centroids[:,0],centroids[:,1]])
center, width, height, phi = lsqe.parameters()

plt.close('all')
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111)
ax.axis('equal')
ax.plot(centroids[:,0], centroids[:,1], 'ro', label='test data', zorder=1)

ellipse = Ellipse(xy=center, width=2*width, height=2*height, angle=np.rad2deg(phi),
               edgecolor='b', fc='None', lw=2, label='Fit', zorder = 2)
ax.add_patch(ellipse)

plt.legend()
plt.show()
```

```python
# 続いてcenterを中心にcenterNodesに偏角を振っていく
import cmath
a_ = np.array([2,4])
b_ = np.array([2,1j])
arg = cmath.phase(np.dot(a_,b_) )
arg
```

```python
def calArg(a,central): #centerから見たときのaの偏角を計算
    comp_ = np.dot(a - central, np.array([1,1j]) )
    return cmath.phase(comp_)

routes = np.zeros((0,2)) # (頂点番号, 角度)の順に格納, ここでいう頂点番号はその頂点が辞書のkeyとして何番目に書いてあるかを返す
#centerNodes.dtype()
for centroid in range(int(len(centerNodes))):
    corr_ = cords[int(centerNodes[centroid])]
    arg = calArg(corr_, center)
    routes = np.append(routes, np.array([[int(centerNodes[centroid]), arg]]), axis = 0)
sortRoute = routes[np.argsort(routes[:, 1])]
sortRoute
```

# 続いてこれらの頂点を順に回るようにダイクストラの実装
＊ただ、自作ダイクストラを回すには頂点の名前とArgを一致させないといけない。。

```python
def find_minimum(array, excludeRange):
    thresh = min(array) * excludeRange
    candidate = np.where(array <= thresh)[0] # np.whereだけだとタプルが返ってくるのでタプルのどこを取ってくるのかを書いておかないといけない！！
    #print(candidate)
    min_ = np.random.choice(candidate)
    #print(min_)
    return min_
    
#find_minimum(np.array([3,2,1,2,1]), 1.02)    

def DijkstraAki(g, s, t, excludeRange): # 自分でいじる方　# タイの場合の処理・通ったパスのリターンを追加
    
    d = np.array([np.inf] * g.number_of_nodes()) # step 1
    b = np.array([False] * g.number_of_nodes()) # step 1
    d[s] = 0 # step 2
    #routeList = [[s] for i in range(g.number_of_nodes())]
    routeFrom = np.zeros(g.number_of_nodes())
    #print(routeList[1].append(8)) # リストのappendメソッドはこれを使ったらどこまででも付きまとう。print分の中だけ、とかではない。
    while not b[t]: # step 3
        i = find_minimum((d + (b*1e100)), excludeRange) # step 4  # false に数をかけると、0として扱われる # タイの処理をしたかったらここをfind_minimum関数に変える
        #routeList[i].append(i)
        if np.isinf(d[i]): break # step4
        b[i] = True # step 4
        for j in g.neighbors(i): # step 4
            d_tent = d[j]
            d[j] = min(d[j], d[i] + g.adj[i][j].get('weight', 1)) # step 4
            if d_tent > d[i] + g.adj[i][j].get('weight', 1): #今までの値より早いパスが見つかった場合
                routeFrom[j] = i
    route = []
    goal = t
    count = 0
    #print(routeFrom[163])
    while goal != s or (count > g.number_of_nodes()):
        route.append(routeFrom[goal])
        goal = int(routeFrom[goal])
        count += 1
        #print(route)
    return d[t], route       
    #return d[t], routeFrom   
```

```python
#import functions as func
nodes = list(pos.keys())
#print(nodes)
#DijkstraAki(G2, 139, 288, 1.01)
route = np.zeros(0)
for i in range(int(len(sortRoute[:,0]) - 1)):
    s, t = int(sortRoute[i,0]), int(sortRoute[i + 1,0])
    
    d, routeList = DijkstraAki(G2, s, t, 1.01) # （グラフ、出発地、目的地, 範囲）
    route = np.append(route, np.array([routeList[::-1]]))
    #print(routeList)
route = np.append(route, sortRoute[-1,0])
route
```
